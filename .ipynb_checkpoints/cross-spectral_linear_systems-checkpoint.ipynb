# Multi-input systems cross spectral analysis

Multi input systems cross-spectral analysis to reveal the relative contribution each of the input terms makes to the output.

This script is based on [Emery and Thomson (2001)](https://www.sciencedirect.com/book/9780444507563/data-analysis-methods-in-physical-oceanography) chapter 5.

An example  disentangling compound dry-heat events across various temporal scales is available in:

[Calim Costa et al.(2023)](https://rdcu.be/dm6I2), Multiscale analysis of drought, heatwaves, and compound events in the Brazilian Pantanal in 2019â€“2021" in Theoretical and Applied Climatology. DOI:10.1007/s00704-023-04655-2. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def multi_system(series,f,names):
    """ Multi-input systems cross spectral analysis
        input:
            series = dictionary of time series
            f = sampling rate 
            names = time series names
        output:
            sizes = partial and multiple coherence
            
    """
    from scipy import signal
    
    n = len(series[0])
    Y = np.fft.fft(series[0])

    X1 =  np.fft.fft(series[1])
    X2 =  np.fft.fft(series[2]) 
    # power
    syy  = (Y * np.conj(Y))
    sxx1 = (X1 * np.conj(X1))
    sxx2 = (X2 * np.conj(X2))
    #cross
    sx1x2 = (X1 * np.conj(X2))    
    syx1 = (Y* np.conj(X1))
    syx2 = (Y* np.conj(X2))
    # transfer function
    H1 = syx1/sxx1
    H2 = syx2/sxx2
    #partial coherence
    cohere1 = H1**2 *sxx1/syy
    cohere2 = H2**2 *sxx2/syy
    cohere1 = (cohere1[int(np.ceil( n/2.)):])
    cohere2 = (cohere2[int(np.ceil( n/2.)):])
    #multi coherence
    coherence = (1-((sxx1*sxx2) - (sx1x2*sx1x2))/(sxx1*sxx2))
    coherence = (coherence[int(np.ceil( n/2.)):])

    #plot
    fig = plt.figure(figsize=(5,3.5), dpi=300)
    fig.text(.5,.95,'Partial and Multiple coherence', horizontalalignment = 'center',fontsize =16)

    ax = plt.subplot(1,1,1)
    size1= np.abs(cohere1[f].real*100)
    size2= np.abs(cohere2[f].real*100)
    size3 = np.abs(coherence[f].real*100)
    sizes = [size1,size2,size3 ]
    #print size1,size2,size3
    models = names#('hw', 'spi','both')
    y_pos = np.arange(len(names))
    colors = ['lightskyblue', 'gold', 'lightsalmon']
    
    plt.barh(y_pos, sizes, color =colors, alpha=0.4)
    plt.yticks(y_pos, models)
    #ax.set_xlabel('% in coherence')

    #plt.axis('equal')
    #plt.savefig('exp09_16.png')
    plt.show()
    return sizes
```

```python
# create a time dimension z

z = np.linspace(0,2048,2048)
# create sine and cosine time series
signal_freq =5
#t=np.arange(0,n/sampling_rate,1/sampling_rate)


x1 = np.sin(5*np.pi*signal_freq*z)
#x2 = 3*np.sin(2 * np.pi *signal_freq*z) 
noise = np.sin(7*np.pi*z)+np.cos(2*np.pi*z)
x2 = np.cos(30*np.pi*z)+noise
y = 1*np.cos(2 * np.pi *signal_freq*z) 

series ={}
series[0] = y
series[1] = x1
series[2] = x2

plt.plot(series[0])
plt.plot(series[1])
plt.plot(series[2])

#f


```

```python
sizes =multi_system(series,500, ['x1','x2','x1+x2'])
```
